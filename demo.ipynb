```python
from minigrad import Scalar, draw_graph
```

```python
a = Scalar(1.5)
b = Scalar(-4.0)
c = a**3 / 5
d = c + (b**2).relu()

d.backward()

draw_graph(d)
```

```python
# This has a nice graph, copied from https://github.com/karpathy/micrograd

a = Scalar(-4.0)
b = Scalar(2.0)
c = a + b
d = a * b + b**3
c += c + 1
c += 1 + c + (-a)
d += d * 2 + (b + a).relu()
d += 3 * d + (b - a).relu()
e = c - d
f = e**2
g = f / 2.0
g += 10.0 / f

g.backward()

draw_graph(g)
```
